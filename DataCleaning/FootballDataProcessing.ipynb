```python
data['FTHG2'] = ""
data['FTHG2'] = data['FTHG'] 
data['FTAG2'] = ""
data['FTAG2'] = data['FTAG'] 
```

```python
HomeTable = (data.groupby('HomeTeam').agg({'FTHG':'sum', 'FTAG':'sum','FTHG2':'mean','FTAG2':'mean'}))
AwayTable = (data.groupby('AwayTeam').agg({'FTAG':'sum', 'FTHG':'sum','FTAG2':'mean','FTHG2':'mean'}))
```

```python
print(HomeTable)
```

```python
print(AwayTable)
```

```python
HomeTable.rename(columns={"FTHG": "Home Goals For", 
                     "FTAG":"Home Goals Against","FTHG2":"Mean Home Goals For","FTAG2":"Mean Home Goals Against"}, inplace=True)
AwayTable.rename(columns={"FTAG": "Away Goals For", 
                     "FTHG":"Away Goals Against","FTAG2":"Away Away Goals For","FTHG2":"Mean Away Goals Against"}, inplace=True)

## renaming columns
```

```python
HomeTable['Total Games Played'] = ""
HomeTable.head()
AwayTable['Total Games Played'] = ""
```

```python
HomeTable['Total Games Played'] = 19 ## inserting total games columns
AwayTable['Total Games Played'] = 19 ## make sure value is not in quotes otherwise the total cant be computed cause it would be recognized as a string value
```

## Now we insert 2 rows (total & mean of all columns)

```python
HomeTable.loc['total']= HomeTable.sum()

HomeTable.loc['Mean']= HomeTable.iloc[:18].mean() ## choosing range of rows, otherwise it will compute the mean of all rows including total

AwayTable.loc['total']= AwayTable.sum()

AwayTable.loc['Mean']= AwayTable.iloc[:18].mean() ## choosing range of rows, otherwise it will compute the mean of all rows including total
```
